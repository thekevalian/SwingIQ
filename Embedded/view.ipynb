```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('SAMPLE47.CSV')
df['time (ms)'] = df['time (ms)'] - df['time (ms)'].min()
df.drop(columns=['temperature (C)'], inplace=True)
df
```

```python
samples_times = (df['time (ms)'] - df['time (ms)'].shift(-1))[:-1]
plt.hist(samples_times, bins=100)
plt.ylim(0, 100)
```

```python
plt.figure(figsize=(12,8))
for column in df.columns:
    if column != "time (ms)":
        plt.scatter(df['time (ms)'], df[column], label=column, s=3)
plt.legend()
plt.show()
```
